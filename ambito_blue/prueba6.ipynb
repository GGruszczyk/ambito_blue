```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Funcin para obtener datos con manejo de errores
def obtener_datos(url):
    try:
        resp = requests.get(url)
        resp.raise_for_status()  # Manejo de errores HTTP
        data = resp.json()[1:]
        df = pd.DataFrame(data, columns=['Date', 'Blue'])
        df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
        df['Da'] = df['Date'].dt.day
        df['Mes'] = df['Date'].dt.month
        df['Ao'] = df['Date'].dt.year
        df.rename(columns={'Blue': 'Valor'}, inplace=True)
        return df
    except requests.exceptions.RequestException as e:
        print(f"Error en la solicitud HTTP: {str(e)}")
        return None
    except Exception as e:
        print(f"Error al obtener datos: {str(e)}")
        return None

# Funcin para graficar datos con personalizacin
def graficar_datos(df, titulo='', etiqueta_x='', etiqueta_y='', guardar_imagen=False):
    if df is not None:
        plt.rcParams['figure.dpi'] = 180
        
        # Configuracin de anotaciones
        anns = {datetime.now(): 'Ud. est aqu'}
        
        plt.figure(figsize=(12, 6))
        plt.plot(df['Date'], df['Valor'], label='Valor del Dlar Informal', color='blue')
        plt.xlabel(etiqueta_x if etiqueta_x else 'Fecha')
        plt.ylabel(etiqueta_y if etiqueta_y else 'Valor')
        plt.title(titulo if titulo else 'Valor del Dlar Informal en Argentina')
        plt.grid(True)
        plt.xticks(rotation=45)
        
        # Anotaciones en el grfico
        for date, label in anns.items():
            plt.annotate(label, xy=(date, df['Valor'].max()), xytext=(date, df['Valor'].max() + 10),
                         arrowprops=dict(facecolor='white', arrowstyle='->', lw=2), color='black', fontsize=8,
                         bbox=dict(boxstyle='square,pad=0.3', fc='white', ec='none', alpha=0.7))
            
        #guardar_imagen

        # plt.tight_layout()
        # plt.legend()
        
        # if guardar_imagen:
        #     fecha_actual = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
        #     nombre_imagen = f'grafico_dolar_{fecha_actual}.png'
        #     plt.savefig(nombre_imagen)
        #     print(f"Grfico guardado como '{nombre_imagen}'")
        # else:
        #     plt.show()

# Construir la URL dinmicamente con la fecha actual
fecha_actual = datetime.now().strftime('%Y-%m-%d')
url = f'https://mercados.ambito.com//dolar/informal/grafico/2000-03-21/{fecha_actual}'

# Obtener y graficar los datos con personalizacin
datos_df = obtener_datos(url)
if datos_df is not None:
    graficar_datos(datos_df, 'Valor del Dlar Informal en Argentina', 'Fecha', 'Valor', True)
```
